# Trading with Leverage and the Kelly Criterion

$Kelly = \frac{\text{Mean return}}{\text{Variance of return}}$

> __The optimal defree of leverage can maximize income/return__

__Kelly vs. sharp ratio__:
- $Kelly = \frac{\text{Reward}}{\text{Risk}^2}$

- $Sharp = \frac{\text{Reward}}{\text{Risk}}$

__The Kelly Criteriom assume that:__
1. Simple returns are used
2. dataset is sufficiently large
3. normaliry of returns for small datasets
4. keep Leverage constant by frequenst trading (its unrealistic)

# Finding the optimal degree of Leverage

```python
simple
```

We can either use __kelly criterion__ or we just __run the backtest for many different leverage settings__.

```python
# set a range of possible leverages 
leverage = np.arange(1, 5, 0.01)
leverage
```

```python
multiple = []
for lever in leverage: # loop on the leverages
    
    # adjust the return to the leverge
    levered_returns = simple[symbol].mul(lever)
    # when leverege small than -1, i set -1 (its imposible to loss more than 100% from the money)
    levered_returns = pd.Series(np.where(levered_returns < -1, -1, levered_returns))
    # compute the compund return, with base of 1    
    multiple.append(levered_returns.add(1).prod())

results = pd.DataFrame(data = {"Leverage":list(leverage), "Multiple":multiple})
```

```python
results.set_index("Leverage", inplace = True)
```

```python
results
```

```python
# get the minimum multiple
results.min()
```

```python
max_multiple = results.max()
max_multiple
```

```python
# This is the optimal leverage
optimal_lev = results.idxmax()
optimal_lev
```

```python
results.plot(figsize = (12, 8));
plt.scatter(x = optimal_lev, y = max_multiple, color = "r", s = 50)
plt.xlabel("Leverage", fontsize = 13)
plt.ylabel("Multiple", fontsize = 13)
plt.title("The optimal degree of Leverage", fontsize = 15)
plt.show()
```

# The impact of Leverage on Reward & Risk

> We are going to calculate  mean return, the standard deviation of returns and the Sharp ratio __for different digrees of leverage.__

> __Importnt Note__ __`We must compute the reward & risk on log return!` otherwise, it'ill be misleading!__

```python
# simple returns
simple
```

 __Reward: 1) Mean of Simple Returns (INCORRECT!)__ 
> we have seen before that there are some problems with the meaning of simple returns, and it's one more time just to demonstrate this.

```python
# Define range of leverages
leverage = np.arange(1, 5, 0.01)
```

```python
# Calculate the mean, std, and sharp ratio for all the different leverages.

# initialize lists
mu = []
sigma = []
sharpe = []

# loop over all the leverages
for lever in leverage:
    # mulruply the returns with the leverage
    levered_returns = simple[symbol].mul(lever)
    # set return that < -1 to be -1 (make sure we can't loos more than 100%)
    levered_returns = pd.Series(np.where(levered_returns < -1, -1, levered_returns))
    
    mu.append(levered_returns.mean()) # mean of simple returns
    sigma.append(levered_returns.std()) # std
    sharpe.append(levered_returns.mean() / levered_returns.std()) # sharp ratio

results = pd.DataFrame(data = {"Leverage":list(leverage),
                                "Mean": mu, 
                                "Std": sigma, 
                                "Sharpe": sharpe})
```

```python
results.set_index("Leverage", inplace = True)
```

```python
results
```

__Plot the mean, std, sharp ratio for all the leverages__

- obviously the __mean of simple leverage return__ is just steadily increasing, __and this pretty much contradicts our previous finding that the multiple__. (where we found the optimal leverage).
- also the sharp ratio stay constant over the time, __and its also misleading__!

> __Its not correct.__

```python
results.plot(subplots = True, figsize = (12, 8), fontsize = 12);
plt.show()

# obviously the mean of simple leverage return is just steadily increasing,
# and this pretty much contradicts our previous finding that the multiple.

# Its clearly misleading!! 
```

__Mean of simple Returns is steadily increasing with higher leverage -> misleading__

__Sharpe Ratio remains constant -> misleading__

..........

 __Reward: 2) Mean of Log Returns (CORRECT)__ 

```python
leverage = np.arange(1, 5, 0.01)
```

```python
# compute the mean, std, and sharp ratio for each leverage level. 
mu = []
sigma = []
sharpe = []

# For each leverage
for lever in leverage:
    # compute the leveraged return
    levered_returns = simple[symbol].mul(lever)
    # set -1 where the return < -1 (we cant loos more then 100%)
    levered_returns = pd.Series(np.where(levered_returns < -1, -1, levered_returns))
    # convert the levereged return to log return (importent!)
    levered_returns = np.log(levered_returns + 1) # convert to log returns
    mu.append(levered_returns.mean()) # mean of log returns
    sigma.append(levered_returns.std()) # std
    sharpe.append(levered_returns.mean() / levered_returns.std()) # sharp ratio

results = pd.DataFrame(data = {"Leverage":list(leverage), 
                               "Mean": mu, 
                               "Std": sigma, 
                               "Sharpe": sharpe})
```

```python
results.set_index("Leverage", inplace = True)
```

```python
results
```

__Plot the mean return, std and sharp ratio of each leverages based on log return!__


```python
results.plot(subplots = True, figsize = (12, 8), fontsize = 12);
plt.show()
```

__`CRITICAL NOTES`__:
- __Maximum Return @ Kelly__
- __Sharpe Ratio steadily decreasing with higher leverage__.
- __Leverage amplifies losses more than it amplifies gains__.
- __Dont use leverage if your goal is to maximize risk-adjusted return__
- __If you want to increase return/income with leverage -> Trade-off__
- __Rule of Thumb: Leverage shouldnt be higher than "Half Kelly".__ 

> __for Kelly critiria, its importent that the log return will be approximatly normal distributed, and will be lot's of data.__

## Putting everything together

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8")
```

```python
returns = pd.read_csv("../Data/returns.csv", index_col = "Date", parse_dates = ["Date"])
returns
```

```python
# compute the kelly criterion
def kelly_criterion(series): # assuming series with log returns
    
    # apply the exponent on log return seruies
    series = np.exp(series) - 1
    if series.var() == 0:
        return np.nan
    else:
        # compute the kelly criterion ratio. 
        return series.mean() / series.var()
```

```python
# now let's apply the kelly_criterion for all the instruments
returns.apply(kelly_criterion).sort_values(ascending = False)
```

```python
# by the kelly criterion, the optimal leverage for the 'Low_vol' instrument
# is leverage of 8.949 
kelly_criterion(returns.Low_Vol)

# BUT.........
```

Side Note: For "Low_Vol", Kelly is not a good approximation because:

- only very few "real" observations (non-normal)
